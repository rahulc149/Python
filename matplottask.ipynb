```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.array([1,10])
b=np.array([2,20])
plt.plot(a,b)
plt.show()
```

```python
plt.plot(a,b,'*')
plt.show()
```

```python
x=np.array([1,3,5,2,6,9,4,22])
y=np.array([2,4,6,8,10,44,3,5])
plt.plot(x,y,marker='*')
plt.show()
```

```python
plt.plot(x,y,marker="o",ms=20)
plt.show()
```

```python
plt.plot(x,y,'o',mec='r')
plt.show()
```

```python
plt.plot(x,y,'o',mfc='r')
plt.show()
```

```python
plt.plot(x,y,'o',linestyle='dashdot')
plt.show()
```

```python
plt.plot(x,y,marker='o',linestyle='dashed',c='blue')
plt.show()
```

```python
plt.plot(x,y,marker='o',linestyle='dashed',c='orange')
plt.show()
```

```python
plt.plot(x,y,marker='*',linestyle='dashed',c='blue',linewidth='10')
plt.show()
```

```python
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("x and y values")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.plot(x,y)
plt.grid()
plt.title("x and y values")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.plot(x,y)
plt.grid(axis='x')
plt.title("x and y values")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
